```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Chargement des données brutes contenant les avis
df_raw = pd.read_csv('../data/olist_clustering.csv')

# 2. Création du graphique
plt.figure(figsize=(9, 5))
sns.countplot(data=df_raw, x='review_score_mean', palette='viridis', edgecolor='black')

plt.title("Graphique 1 : Distribution brute des scores d'avis clients", fontsize=13, fontweight='bold', pad=15)
plt.xlabel("Note de l'avis (Review Score de 1 à 5)", fontsize=11)
plt.ylabel("Nombre de clients", fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. Chargement des données 
df_raw = pd.read_csv('../data/olist_clustering.csv')

# 2. Création du graphique corrigé (Histogramme)
plt.figure(figsize=(9, 5))
sns.histplot(
    data=df_raw, 
    x='review_score_mean', 
    bins=[1, 2, 3, 4, 5, 6], # Définit des intervalles nets de 1 en 1
    discrete=False, 
    kde=False, 
    color='teal', 
    edgecolor='black',
    alpha=0.8
)

# Ajustement de l'axe X pour qu'il soit propre
plt.xticks([1.5, 2.5, 3.5, 4.5, 5.5], labels=['[1 - 2[', '[2 - 3[', '[3 - 4[', '[4 - 5[', '5 / 5'])

plt.title("Graphique 1 : Distribution brute de la satisfaction client (Données Originales)", fontsize=13, fontweight='bold', pad=15)
plt.xlabel("Intervalles des notes moyennes de satisfaction", fontsize=11)
plt.ylabel("Nombre de clients (Effectifs)", fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# 1. Sélection des variables après Feature Engineering
features_engineered = ['recency', 'frequency', 'monetary_value', 'mean_freight', 'review_score_mean']
corr_matrix = df_raw[features_engineered].corr()

# 2. Création de la Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, vmin=-1, vmax=1)

plt.title("Graphique 2 : Matrice de corrélation des Features standardisées", fontsize=13, fontweight='bold', pad=15)
plt.tight_layout()
plt.show()
```
